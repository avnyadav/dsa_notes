```python
def dfs(node,parent,timer,discovery,low,result,adjList,visited):
    visited[node]=True
    discovery[node],low[node]=timer,timer
    timer+=1
    for neigh in adjList[node]:
        if neigh==parent:
            continue
        #print(visited,neigh)
        if not visited[neigh]:
            dfs(neigh,node,timer,discovery,low,result,adjList,visited)
            low[node]=min(low[node],low[neigh])

            #check for edge

            if low[neigh]>discovery[node]:
                ans= []
                ans.append(node)
                ans.append(neigh)
                result.append(ans)
        else:
            #back edge
            low[node] = min(low[node],discovery[neigh])








def findBridges(edges, v, e):

    # Write your code gere
    #print("vertices",v)
    adjList= { i:list() for i in range(v)}
    for edge in edges:
        startNode,stopNode= edge[0],edge[1]
        adjList[startNode].append(stopNode)
        adjList[stopNode].append(startNode)
    #print(adjList)
    timer=0
    discovery = [-1]*v
    low = [-1]*v
    parent=-1
    visited=[False]*v
    #print(visited)
    result= []

    for i in range(v):
        if not visited[i]:
            dfs(i,parent,timer,discovery,low,result,adjList,visited)

    return result
```

```python
from os import *
from sys import *
from collections import *
from math import *
from typing import List

"""
            col0    col1    col2    col3
row 0
row 1
row 2
row 3



"""

def isPlacementPossible(row,col,currentPlacement,n):

    temp_col=col
    #check for row
    #to skip current column
    while temp_col>0:
        temp_col=temp_col-1
        if currentPlacement[row][temp_col]==1:
            return False
        
    
    #diagonal left upward check
    temp_col=col
    temp_row=row
    while temp_col>0 and temp_row>0:
        temp_col=temp_col-1
        temp_row=temp_row-1
        if currentPlacement[temp_row][temp_col]==1:
            return False
        


  #diagonal left upward check
    temp_col=col
    temp_row=row
    while temp_col>0 and temp_row<n-1:
        temp_col=temp_col-1
        temp_row=temp_row+1
        if currentPlacement[temp_row][temp_col]==1:
            return False

    return True



def placeNQueens(col:int,n:int,currentPlacement,ans):

    #exit criteria
    if col==n:
        temp=[]
        for row in currentPlacement:
            for queenPosition in row:
                temp.append(queenPosition)
        ans.append(temp)
        return
    
    for row in range(n):
        
        if isPlacementPossible(row,col,currentPlacement,n):
            
            currentPlacement[row][col]=1
            placeNQueens(col+1,n,currentPlacement,ans)
            currentPlacement[row][col]=0
    

def nQueens(n):

    # Write your code here
    currentPlacement=[]
    for i in range(n):
        currentPlacement.append([])
        for j in range(n):
            currentPlacement[i].append(0)
    
    ans=[]

    placeNQueens(0,n,currentPlacement,ans)

    return ans
```

```python
import heapq


class MaxHeap:

	def __init__(self):
		self.arr=[]
	
	def push(self,item):
		heapq.heappush(self.arr,-1*item)
	
	def pop(self):
		return -1*heapq.heappop(self.arr)
	
	@property
	def size(self):
		return len(self.arr)
	
	def top(self):
		if self.size>0:
			return -1*self.arr[0] 


class MinHeap:

	def __init__(self):
		self.arr=[]
	
	def push(self,item):
		heapq.heappush(self.arr,item)
	
	def pop(self):
		return heapq.heappop(self.arr)
	
	@property
	def size(self):
		return len(self.arr)
	
	def top(self):
		if self.size>0:
			return self.arr[0]




def signam(maxHeap,minHeap):
    if maxHeap.size-1==minHeap.size:
        return 1
    if maxHeap.size==minHeap.size-1:
        return -1
    else:
        return 0

def equalHeap(maxHeap,minHeap,element,medianValue):
    if element<medianValue:
        maxHeap.push(element)
        medianValue=maxHeap.top()
    else:
        minHeap.push(element)
        medianValue=minHeap.top()
    return medianValue

def getMedianValueIfSizeEqual(maxHeap,minHeap):
    #Equal size i.e. take average of middle
    return (maxHeap.top()+minHeap.top())//2


def largeMaxHeap(maxHeap,minHeap,element,medianValue):
    if element<medianValue:
        
        topMaxHeapEle = maxHeap.top()
        minHeap.push(topMaxHeapEle)
        maxHeap.pop()
        maxHeap.push(element) 
        #after this step both side are equal
        medianValue = getMedianValueIfSizeEqual(maxHeap,minHeap)

    else:
        minHeap.push(element)
        medianValue=getMedianValueIfSizeEqual(maxHeap,minHeap)
    return medianValue

def largeMinHeap(maxHeap,minHeap,element,medianValue):
    if element<medianValue:
        maxHeap.push(element)
        medianValue=getMedianValueIfSizeEqual(maxHeap,minHeap)
    else:
        topMinHeapEle = minHeap.top()
        maxHeap.push(topMinHeapEle)
        minHeap.pop()
        minHeap.push(element)
        medianValue=getMedianValueIfSizeEqual(maxHeap,minHeap)
    return medianValue



def nextMedian(element,maxHeap,minHeap,medianValue):
    if medianValue is None:
        medianValue=element
        maxHeap.push(element)
        return medianValue
    caseNum = signam(maxHeap=maxHeap,minHeap=minHeap)
    if caseNum==0:
        medianValue=equalHeap(maxHeap=maxHeap,
                                                minHeap=minHeap,
                                                element=element,
                                                medianValue=medianValue)
    
    if caseNum==1:
        medianValue = largeMaxHeap(maxHeap=maxHeap,
                                                minHeap=minHeap,
                                                element=element,
                                                medianValue=medianValue)
    if caseNum==-1:
        medianValue = largeMinHeap(maxHeap=maxHeap,
                                                minHeap=minHeap,
                                                element=element,
                                                medianValue=medianValue)
    
    return medianValue


def findMedian(arr, n):
    maxHeap = MaxHeap()
    minHeap = MinHeap()
    medianValues = []
    medianValue=None
    for ele in arr:
        medianValue=nextMedian(element=ele,maxHeap=maxHeap,minHeap=minHeap,medianValue=medianValue)
        medianValues.append(medianValue)
    # Write your code here
    return medianValues
```

```python
from typing import List
```

```python
class TrieNode:


    def __init__(self,data:str) -> None:
        self.data:str=data
        self.children:List[TrieNode]=[]
        for i in range(26):
            self.children.append(None)
        self.isTerminal:bool=False
```

```python
class Tries:


    def __init__(self) -> None:
        self.root:TrieNode=TrieNode(data=None)

    def insertUtil(self,root:TrieNode,word):
        if len(word)==0:
            root.isTerminal=True
            return
        #Assuming All word will be in capital letter
        letter=word[0]
        
        if root.children[ord(letter)-ord('A')] is None:
            #absent
            childNode = TrieNode(data=letter)
            root.children[ord(letter)-ord('A')]=childNode
        else:
            #present
            childNode=root.children[ord(letter)-ord('A')]
        self.insertUtil(childNode,word=word[1:])


    def insertWord(self,word):
        self.insertUtil(self.root,word)

    def searchUtil(self,root:TrieNode,word):
        if len(word)==0:
            return root.isTerminal
        letter = word[0]

        if root.children[ord(letter)-ord('A')] is not None:
            childNode = root.children[ord(letter)-ord('A')]
        else:
            return False
        return self.searchUtil(childNode,word[1:])
    def search(self,word):
        return self.searchUtil(root=self.root,word=word)
    
    def removeUtil(self,root:TrieNode,word:str):
        if len(word)==0:
            root.isTerminal=False
            return
        letter = word[0]
        
        if root.children[ord(letter)-ord('A')] is None:
            return
        else:
            childNode = root.children[ord(letter)-ord('A')]

         
        self.removeUtil(root=childNode,word=word[1:])

    def removeWord(self,word):
        self.removeUtil(root=self.root,word=word)
```

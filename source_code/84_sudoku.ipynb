```python
from os import *
from sys import *
from collections import *
from math import *

def isNumberAcceptable(row,col,number,sudoku):
    #check if a row contain same number already
    for c in range(9):
        if sudoku[row][c]==number:
            return False
        if sudoku[c][col]==number:
            return False
         
    #sub cube
    
    start_col_idx=(col//3)*3
    start_row_idx=(row//3)*3
    for r in range(start_row_idx,start_row_idx+3):
        for c in range(start_col_idx,start_col_idx+3):
            if sudoku[r][c]==number:
                return False
    return True

def solve(sudoku):
    skip=False
    solved=True
    for r in range(0,9):
        for c in range(0,9):
            skip=False
            if sudoku[r][c]==0:
                solved=False
                #try each number
                for num in range(1,10):
                    if skip:
                        continue
                    numberAccepted = isNumberAcceptable(r,c,num,sudoku)
                    if numberAccepted:
                        #proceed with next position
                        sudoku[r][c]=num
                        solved = solve(sudoku)
                        if not solved:
                            sudoku[r][c]=0
                        skip=solved
                return solved
                

    return solved            

        
 
def solveSudoku(sudoku):
  
    
    solve(sudoku)
   
    # Write your code here.

if __name__=="__main__":
    data="""3 0 6 5 0 8 4 0 0
5 2 0 0 0 0 0 0 0
0 8 7 0 0 0 0 3 1
0 0 3 0 1 0 0 8 0
9 0 0 8 6 3 0 0 5
0 5 0 0 9 0 6 0 0
1 3 0 0 0 0 2 5 0
0 0 0 0 0 0 0 7 4
0 0 5 2 0 6 3 0 0"""
    sudoku=[]
    i=0
    for row in data.split("\n"):
        sudoku.append(list())
        for col in map(int,row.split(" ")):
            #print(col,end="\t")
            sudoku[i].append(col)
        i=i+1
        print("\n")
    #print(sudoku)
    solveSudoku(sudoku)
    print(sudoku)
```

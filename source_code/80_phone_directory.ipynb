```python
accepted_data="abcdefghijklmnopqrstuvwxyz"
from typing import List
class TrieNode:


    def __init__(self,data:str) -> None:
        self.data:str=data
        self.children:List[TrieNode]=[]
        for i in range(26):
            self.children.append(None)
        self.isTerminal:bool=False



class Trie :

    def __init__(self) :
        self.root:TrieNode=TrieNode(data=None)

        # constructor for the Trie
        pass

    def insertUtil(self,root:TrieNode,word):
        if len(word)==0:
            root.isTerminal=True
            return
        #Assuming All word will be in capital letter
        letter=word[0]
      
        
        if root.children[ord(letter)-ord('a')] is None:
            #absent
            childNode = TrieNode(data=letter)
            root.children[ord(letter)-ord('a')]=childNode
        else:
            #present
            childNode=root.children[ord(letter)-ord('a')]
        self.insertUtil(childNode,word=word[1:])

    def insert(self, string) :
        self.insertUtil(self.root,string)

    def printSuggestion(self,node,temp,prefix):
        if node.isTerminal:
            temp.append(prefix)
        for child in node.children:
            if child is not None:
                self.printSuggestion(node=child,temp=temp,prefix=f"{prefix}{child.data}")

    

        

    def getSuggestion(self,string):
        string=string.strip()
        prevNode = self.root
        output=[]
        prefix=""
        for letter in string:
          
            prefix=  f"{prefix}{letter}"
            if letter not in accepted_data:
                raise Exception(f"Testing {string} {prefix} [{letter}] {len(letter)}")
            node = prevNode.children[ord(letter)-ord('a')]
            if node is None:
                return output
            temp=[]
            self.printSuggestion(node,temp,prefix)
            output.append(temp)
            prevNode=node
        return output



def phoneDirectory(arr,qList):
    t = Trie()
    
    for ele in arr:
        t.insert(ele)
    return t.getSuggestion(string=qList)
```

```python
from os import *
from sys import *
from collections import *
from math import *

# def updateMap(dictionary,k,v):
#     if not dictionary.get(k,None):
#         dictionary[k]=[v]
#         return
#     dictionary[k].append(v)

def prepareAdjList(adjList,edges,e):

    for edge in edges[:e]:
        u,v=edge[0],edge[1]
        if u is not None:
            adjList[u].append(v)



def topologicalSort(adj, v, e):
    # Write your code here

    adjList=dict(zip([i for i in range(v)],[list() for _ in range(v)]))


    prepareAdjList(adjList,adj,e)
    #print(adjList)

    #preparing indegree

    indegree=dict()

    for key in adjList.keys():
        for node in adjList[key]:
            if not indegree.get(node,False):
                indegree[node]=1
            else:
                indegree[node]+=1
    
    #insert all node with indegree 0
    queue=[]
    for i in range(v):
        if not indegree.get(i,False):
            indegree[i]=0
        if indegree[i]==0:
            queue.append(i)

    ans=[]
    while len(queue)!=0:
        front = queue.pop(0)
        ans.append(front)

        for neighbour in adjList[front]:
            indegree[neighbour]-=1
            if indegree[neighbour]==0:
                queue.append(neighbour)



    return ans

    
```

```python

def prepareAdjList(adjList,edges,e):

    for edge in edges[:e]:
        u,v=edge[0],edge[1]
        if u is not None:
            adjList[u].append(v)


def detectCycleInDirectedGraph(n, edges):
    # Write your code here
    adjList=dict()

    #print(adjList)
    
    adjList=dict(zip([i for i in range(1,n+1)],[list() for _ in range(1,n+1)]))


    prepareAdjList(adjList,edges,len(edges))
    #print(adjList)

    #preparing indegree

    indegree=dict()

    for key in adjList.keys():
        for node in adjList[key]:
            if not indegree.get(node,False):
                indegree[node]=1
            else:
                indegree[node]+=1
    
    #insert all node with indegree 0
    queue=[]
    for i in range(1,n+1):
        if not indegree.get(i,False):
            indegree[i]=0
        if indegree[i]==0:
            queue.append(i)
    cnt=0
    while len(queue)!=0:
        front = queue.pop(0)
        #ans.append(front)
        cnt+=1
        for neighbour in adjList[front]:
            indegree[neighbour]-=1
            if indegree[neighbour]==0:
                queue.append(neighbour)
     
    
    return not cnt==n
    
```

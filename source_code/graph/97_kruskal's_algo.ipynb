```python
from sys import stdin,setrecursionlimit
setrecursionlimit(10**7)



# Edge class for storing the Edges of thee graph
class Edge:
    
    def __init__(self, start, end, weight) :

        self.start = start
        self.end = end
        self.weigth = weight

    def __lt__(self,other):
        return self.weigth<other.weigth


def findParent(parent,node):
    if parent[node]==node:
        return node
    parent[node]=findParent(parent,parent[node])
    return parent[node]


def unionSet(u,v,parent,rank):
    u = findParent(parent,u)
    v = findParent(parent,v)

    if rank[u]<rank[v]:
        parent[u]=v
    elif rank[v]<rank[u]:
        parent[v]=u
    else:
        parent[u]=v
        rank[v]+=1
    


def minimumSpanningTree(edges, V, E):
    #print(edges,V,E)
	# Your code goes here
    # return your answer
    parent=[ i for i in range(V)]
    rank=[0 for i in range(V)]
    edges.sort()

    total=0
    for i in range(len(edges)):
        u = edges[i].start
        v=edges[i].end
        w = edges[i].weigth

        u = findParent(parent,u)
        v=findParent(parent,v)

        if u!=v:
            total+=w
            unionSet(u,v,parent,rank)


    return total































#taking inpit using fast I/O
def takeInput() :
    n_m = stdin.readline().strip().split(" ")
    V = int(n_m[0].strip())
    E = int(n_m[1].strip())

    edges = [Edge(0, 0, 0) for i in range(E)]
    for i in range(E) :

        temp = list(map(int, stdin.readline().strip().split(" ")))
        edges[i] = Edge(temp[0], temp[1], temp[2])

    return edges, V, E


#main
edges, V, E = takeInput()
print(minimumSpanningTree(edges, V, E))
```

```python
from os import *
from sys import *
from collections import *
from math import *


def prepareAdjList(n,m,g):
    adjList=dict()
    for i in range(1,n+1):
        adjList[i]=list()

    for e in g:
        u,v,w=e[0],e[1],e[2]
        adjList[u].append((v,w))
        adjList[v].append((u,w))
    return adjList

def getInitalValue(n,val):
    var=dict()
    for i in range(1,n+1):
        var[i]=val
    return var


def getMinimumWeight(mst,weight):

    min_val=None
    ans = None
    for key,val in weight.items():
        if mst[key] is False and  min_val is None:
            min_val=val
        if mst[key] is False and val<=min_val:
            ans=(key,val)
            min_val=val
    
    return ans




def calculatePrimsMST(n, m, g):
    # Write your code here.
    # Return a 2-D list.
    ans=[]
    adjList = prepareAdjList(n,m,g)
    
    mst,key,parent = dict(),dict(),dict()
    for i in range(1,n+1):
        mst[i]= False
        key[i]= float('inf')
        parent[i]= -1

    key[1]=0
    parent[1]=-1

    for i in range(1,n):
        mini=float('inf')
        u=None
        #find  mini node
        for v in range(1,n+1):
            if mst[v]==False and key[v]<mini:
                u=v
                mini=key[v]
        if u is None:
            break
        mst[u]=True

        for neigh in adjList[u]:
            v=neigh[0]
            w=neigh[1]
            if mst[v]==False and  w<key[v]:
                parent[v]=u
                key[v]=w

    ans=[]
    for i in range(2,n+1):
        u,v,w=parent[i],i,key[i]
        ans.append([u,v,w])

    return ans
```
